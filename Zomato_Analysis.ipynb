# Handling Null Values

```python
df.isnull().sum()
```

```python
df['cuisines'].fillna(df['cuisines'].mode()[0], inplace=True)

df['aggregate_rating'] = pd.to_numeric(df['aggregate_rating'], errors='coerce')

df['votes'] = pd.to_numeric(df['votes'], errors='coerce')

df['price_range'] = pd.to_numeric(df['price_range'], errors='coerce')

```

```python
df['timings'].fillna("Not Available", inplace=True)

df['address'].fillna( "Unknown",inplace=True)

df['opentable_support'].fillna(df['opentable_support'].median(), inplace=True)
```

```python
df.drop(["zipcode"],axis = 1, inplace = True)
```

## Fill categorical missing values

```python
categorical_cols = ['cuisines', 'currency', 'rating_text']

for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

## Fill numerical missing values

```python
numerical_cols = [
    'aggregate_rating',
    'votes',
    'price_range',
    'photo_count'
]

for col in numerical_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

# Feature engineering

```python
df['restaurant_type'] = pd.cut(df['average_cost_for_two'], bins=[0,500,1500,3000,10000], labels=['Budget','Mid-Range','Premium','Luxury'])

df['high_rated'] = np.where(df['aggregate_rating'] >= 4,'Yes','No')

df['number_of_cuisines'] = (df['cuisines'].astype(str).apply(lambda x: len(x.split(','))))
```

```python
df['delivery_available'] = df['delivery'].map({1:'Yes',0:'No'})

df['engagement_level'] = pd.cut(df['votes'], bins=[0,50,200,500,10000], labels=['Low','Medium','High','Very High'])

df['chain_restaurant'] = (df['name'].duplicated(keep=False).map({True:'Yes', False:'No'}))
```

```python
df['good_value'] = np.where((df['aggregate_rating'] >= 4) &(df['average_cost_for_two'] <= 1000),'Yes','No')

df['popularity_score'] = (df['aggregate_rating'] + df['votes'] * 0.01)
```

### df["restaurant_type"].fillna("Not Recorded", inplace = True)

df["engagement_level"].fillna("No records", inplace = True)

Your syntax is correct, but there is a small issue:

If restaurant_type and engagement_level are created using pd.cut(), they become categorical columns.
For categorical columns, filling a new value directly may sometimes raise this error:

Cannot setitem on a Categorical with a new category

```python
df["restaurant_type"] = (df["restaurant_type"].astype(str))
df["restaurant_type"].fillna("Not Recorded",inplace=True)

df["engagement_level"] = (df["engagement_level"].astype(str))
df["engagement_level"].fillna("No Records",inplace=True)
```

```python
df["delivery_available"].fillna("Not mention", inplace = True)
```
